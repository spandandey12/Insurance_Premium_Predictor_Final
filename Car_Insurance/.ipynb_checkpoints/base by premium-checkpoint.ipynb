```python
import pandas as pd
df = pd.read_csv('/Users/spandandey/Desktop/courses/Data 533/insurance/car_insurance_premium_dataset.csv')
df.head()
```

```python
for col in df.columns:
    print(f"\n Columns: {col}")
    print("-------------------")
    print(f"\n Unique values: {df[col].unique()}")
    #print("")
    print(f"\n Value Count: {df[col].value_counts(dropna=False)}")
    #print("")
    print(f"\n Percentage: {df[col].value_counts(dropna=False,normalize=True).round(2)}")

    print("-------------------")
```

```python
age_factor_mean = df.groupby("Driver Age")["Insurance Premium ($)"].mean()
insurance_premium_mean = df["Insurance Premium ($)"].mean()
age_multi = age_factor_mean/insurance_premium_mean
print(age_multi)
```

```python
driver_experience_mean = df.groupby("Driver Experience")["Insurance Premium ($)"].mean()
drive_experience_multi = driver_experience_mean/insurance_premium_mean
print(drive_experience_multi)
```

```python
previous_accident_mean = df.groupby("Previous Accidents")["Insurance Premium ($)"].mean()
previous_accident_multi = previous_accident_mean/ insurance_premium_mean
print(previous_accident_multi)
```

```python
annual_mileage_mean = df.groupby("Annual Mileage (x1000 km)")["Insurance Premium ($)"].mean()
annual_mileage_multi = annual_mileage_mean / insurance_premium_mean

print(annual_mileage_multi)
```

```python
car_age_mean = df.groupby("Car Age")["Insurance Premium ($)"].mean()
car_age_multi = car_age_mean / insurance_premium_mean
print(car_age_multi)
```
